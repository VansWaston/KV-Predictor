# 0. Import

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

# 1. 生成数据集

```python
# 生成数据集：1000个数据点，3个簇
X, _ = make_blobs(n_samples=1000, centers=3, random_state=42)
```

```python
# 可视化原始数据
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=10, c='blue')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# 2. DBSCAN并尝试三组参数

```python
# 尝试三组不同的 DBSCAN 参数组合
params = [(0.3, 10), (0.3, 15), (0.5, 15), (0.7, 20)]
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

for (eps, min_samples), ax in zip(params, axes.flatten()):
    # 使用 DBSCAN 聚类
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    
    # 可视化聚类结果
    scatter = ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=10)
    ax.set_title(f'DBSCAN Clustering (eps={eps}, minPts={min_samples})')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
